```python
fp = flight_price.pivot(index='id', columns=['departureAirport','arrivalAirport'], values='totalPrice').fillna(0)
fp
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
def findSimilarPrice(flight, fp):
    similarity = []
    for i,row in enumerate(fp.values):
        cos = cosine_similarity(flight, row.reshape(1,-1))[0][0]
        similarity.append([i, cos])
    temp = pd.DataFrame(similarity, columns=['id', 'similarity'])
    temp = temp.sort_values(by='similarity', ascending=False)
    similar_flights = list(temp['id'].values)
    similarities = list(temp['similarity'].values)
    return similar_flights, similarities
```

```python
test_flight = fp.iloc[0].values.reshape(1,-1)
print(test_flight)
test_recs = findSimilarPrice(test_flight, fp)
```

```python
def findSimilarFlights(flight, fp, recommendations=10, analyzed_flights=10):
    seen = list(fp.columns[list(flight>0)[0]])
    print("Seen: ", seen)
    similars = findSimilarPrice(flight, fp)
    print("Similars: ", similars)

    scores = pd.DataFrame(columns=['id', 'score'])
    dtypes = np.dtype([('id', int), ('score', float)])

    for sim_flight, sim_score in zip(similars[0][0:analyzed_flights], similars[1][0:analyzed_flights]):
        rec_flights = fp.iloc[sim_flight]
        #rec_index = list(rec_flights[rec_flights>0].index.values)
        rec_index = rec_flights
        print(rec_index.shape)
        if (len(rec_index) > 0):
            for flight in rec_index:
                print("Flight: ", flight)
                if (flight not in scores['id'].values):
                    scores.loc[len(scores)] = (flight, 10*sim_score)
                else:
                    scores.loc[scores['id']==flight, 'score'] += 10*sim_score
        scores.sort_values(by='score', ascending=False, inplace=True)

        try:
            scores = scores.head(recommendations)
            scores['id'] = scores['id'].astype(int)
        except:
            scores['id'] = scores['id'].astype(int)
    
    return list(scores['id'].values), list(scores['score'].values)
```

```python
ids, scores = findSimilarFlights(test_flight, fp, 10, 10)   
pd.DataFrame({'id':ids, 'score':scores})
```
